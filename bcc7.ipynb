```python
msg = "gasgg54@#vscsd!s*"

special = 0
spaces = 0

for ch in msg:
    if ch == " ":
        spaces += 1
    elif not ch.isalnum():
        special += 1

print("Special:", special)
print("Spaces:", spaces)
print("Total:", special + spaces)
```

```python
#move zeros to the end
#Question: Given an array ,move all the zeros to the end without changing the order of non zero elements
#SampleInput:[0, 1,0,3,12]
#ExpectedOutput:[1,3,12,0,0]
arr = [0, 1, 0, 3, 12]

j = 0  # position for non-zero elements

for i in range(len(arr)):
    if arr[i] != 0:
        arr[j] = arr[i]
        j += 1

# fill remaining positions with 0
for i in range(j, len(arr)):
    arr[i] = 0

print(arr)
```

```python
#Finding the total distance between adjacent terms of a list of 5 numbers
#Input:5
#10 11 7 12 14
#Output:12
arr = list(map(int, input().split()))

total = 0

for i in range(len(arr) - 1):
    total += abs(arr[i] - arr[i + 1])

print(total)
```
